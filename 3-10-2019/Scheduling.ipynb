```python
import pygal
```

```python
Cbar = pygal.StackedBar()
```

```python
Cbar.title = 'CPU Scheduling'
```

```python
Cbar.x_labels = map(str, ('cpu1', 'cpu2','cpu3'))
```

```python
from collections import Counter
from itertools import count

jobs = [1.1,2.1,3,1,0.7,5,3,2,0.9,1.8,4.6,3.1,1.6,0.6,2.8]
jobs.sort(reverse=True)
time1=0
time2=0
time3=0
cpu1=[]
cpu2=[]
cpu3=[]
while len(jobs)!=0:
    #print(jobs)
    if time1==min(time1,time2,time3):
        time1+=jobs[0]/2.66
        cpu3.append(jobs[0])
        o=str(jobs[0])
        Cbar.add(o, [0,0,jobs[0]/2.66])
    elif time2==min(time1,time2,time3):
        time2+=jobs[0]/2
        cpu2.append(jobs[0])
        o=str(jobs[0])
        Cbar.add(o, [0,jobs[0]/2,0])
    elif time3==min(time1,time2,time3):
        time3+=jobs[0]/1.33
        cpu1.append(jobs[0])
        o=str(jobs[0])
        Cbar.add(o, [jobs[0]/1.33,0,0])
    jobs.pop(0)
    
print(max(time1,time2,time3))
print("CPU1:", cpu1)
print("CPU2:", cpu2)
print("CPU3:", cpu3)
```
